```python
#Location frequency
from collections import Counter
print(Counter(df["location.name"]), end = " ")
```

```python
counts = Counter(df["location.name"])
```

```python
#conversion into dictionary
counts = dict(counts)

#appending values of the dictionary in freq variable
freq = []
for key, val in counts.items():
  freq.append(val)

#sorting and reverse sorting for starting with maximum value
freq = sorted(freq)
freq = freq[::-1]
print(freq, end = " ")
```

```python
#sorting dictionary by value
import operator
x = {1: 2, 3: 4, 4: 3, 2: 1, 0: 0}
sorted_x = sorted(counts.items(), key=operator.itemgetter(1))
sorted_x = dict(sorted_x)

names = []
for name in sorted_x.keys():
  names.append(name)
names = names[::-1]
print(names, end= " ")
```

```python
#creating a dataframe with columns- Location and Number of appearances in earthquake.csv datasets

df = pd.DataFrame({"Location" : names, "Numbers" : freq })
df.head()
```

```python
#creating a bar diagram
plt.figure(figsize = (20, 6))
plt.bar(df["Location"], df["Numbers"])
plt.xticks(rotation = 90)
plt.xlabel("Location")
plt.ylabel("Counts")
plt.show()
```

```python
import pandas as pd
import numpy as np

earthquake= pd.DataFrame({
    "PHI_E" : [0,0,0,0,40,40,40,60,60,60],
    "LAMBDA_E" : [0,0,0,-30,0,0,0,0,0,0],
    "PHI_S": [1,5,10,60,41,45,50,61,65,70],
    "LAMBDA_S": [1,5,10,60,1,5,10,1,5,10]

})

delX = np.array(111.195*(earthquake["PHI_S"] - earthquake["PHI_E"]))
delY = np.array(111.195*(earthquake["LAMBDA_S"] - earthquake["LAMBDA_E"])*np.cos((earthquake["PHI_E"] + earthquake["PHI_S"]) / 2))

# distance, azimuth, back azimuth to be included as columns of the earthquake dataset
dist_km = np.sqrt(np.power(delX, 2) + np.power(delY, 2))
azimuth = np.degrees(np.arctan(delY/delX))
backAzimuth = np.degrees(np.arctan(delY/delX)) + 180

earthquake['Dist(km)'] = dist_km
earthquake['Azimuth'] = azimuth
earthquake['backAzimuth'] = backAzimuth
earthquake
```

```python
# epicenter and station coordinates input
# get results

phi_e = int(input("Enter the epicenter latitude : "))
lambda_e = int(input("Enter the epicenter longitude : "))
phi_s = int(input("Enter the station latitude : "))
lambda_s = int(input("Enter the station longitude : "))

delX = 111.195 * (phi_s - phi_e)
delY = 111.195 * (lambda_s - lambda_e) * np.cos((phi_e + phi_s) / 2)

dist_km = np.sqrt(np.power(delX, 2) + np.power(delY, 2))
Az = np.degrees(np.arctan(delY/delX))
Baz = np.degrees(np.arctan(delY/delX)) + 180

print("Epicentral distance {:.2f} km".format(dist_km))
print("Azimuth = {:.2f} degree".format(Az))
print("Back azimuth = {:.2f} degree".format(Baz))
```
