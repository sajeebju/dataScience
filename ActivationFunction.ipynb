```python
import matplotlib.pyplot as plt
import numpy as np
import math

X = np.arange(-10, 10)

def plot(x, func, title, color):
    plt.plot(x, func, lw = 3, c = color)
    plt.axhline(lw = 1, c = 'c')
    plt.axvline(lw = 1, c = 'c')
    #plt.xlim(-10, 10)
    #plt.ylim(-2, 2)
    plt.grid(alpha = 0.4)
    plt.title(title)
    plt.box(on = None)
```

```python
#Sigmoid Function

def sigmoid(z):
    return 1/ (1 + np.exp(-z))

plot(X, sigmoid(X), "Sigmoid Function", "red")
```

```python
#RELU Activation Function

def relu(x):

  x1 = []
  for i in x:

    if i < 0:
      x1.append(0)

    else:

      x1.append(i)

  return x1
```

```python
plot(X, relu(X), "Relu Activation Function", "c")
```

```python
# SOFTMAX Activation Function

def softMax(z):

  return np.exp(z) / np.sum(np.exp(z), axis= 0 )
```

```python
plot(X, softMax(X), "Softmax Activation Function", "r")
```

```python
#Bipolar Sigmoid Function

def bipolar_sigmoid(z):
    return (1 - np.exp(-z)) / (1 + np.exp(-z))
```

```python
plot(X, bipolar_sigmoid(X), "Bipolar Sigmoid Function", "y")
```

```python
#tanh Activation Function

def tanh(x):
  return np.tanh(x)
```

```python
plot(X, tanh(X), "Tanh Activation Function", "k")
```
